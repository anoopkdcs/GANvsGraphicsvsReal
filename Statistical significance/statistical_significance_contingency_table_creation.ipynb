**Creating contingency table for computing statistical significance**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from statsmodels.stats.contingency_tables import mcnemar
from scipy.stats import ttest_ind
from scipy.stats import ks_2samp
from sklearn import metrics
import csv
```

**1. Contingency table for MC-EffNet-2 vs. Best baseline**

```python
file1 = open('path to MC-EffNet-2 prediction csv file')# our model
csvreader = csv.reader(file1)
mceffnet2 = []
for row in csvreader:
  mceffnet2.append(row)

print(len(mceffnet2))

file2 = open('path to best baseline prediction csv file')#InceptionResnet
csvreader = csv.reader(file2)
inception_resnet = []
for row in csvreader:
        inception_resnet.append(row)
print(len(inception_resnet))
```

```python
contingency_table_mceffnet2_with_baseline = metrics.confusion_matrix(mceffnet2, inception_resnet)
print("COntigency table mceffnet with best baseline \n",contingency_table_mceffnet2_with_baseline)
```

**2. Contingency table for MC-EffNet-2 vs. Manual Annotations**

```python
############# Manual annotation significance #############
file3 = open('path to MC-EffNet-2 prediction csv file')#
csvreader = csv.reader(file3)
psycho_mceffnet2 = []
for row in csvreader:
  psycho_mceffnet2.append(row)

print(len(psycho_mceffnet2))

file4 = open('path to manual prediction csv file')#
csvreader = csv.reader(file4)
psycho_manual = []
for row in csvreader:
        psycho_manual.append(row)
print(len(psycho_manual))
```

```python
contingency_table_psycho_mceffnet2_with_manual = metrics.confusion_matrix(psycho_mceffnet2, psycho_manual)
print("COntigency table Psycho Experiments \n",contingency_table_psycho_mceffnet2_with_manual)
```
